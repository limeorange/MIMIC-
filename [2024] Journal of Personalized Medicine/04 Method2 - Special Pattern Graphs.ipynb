# Definition of Visualization Function

```python
# 0. Create a dict containing feature name information for 4069 types of features
feature_df = pd.read_csv('feature_df.csv')
feature_name = dict(zip(feature_df['feature'], feature_df['feature_name']))

X = np.load('x_(7727,10,4068).npy')
y = np.load('/project/LSH/y_(7727,1).npy')
COLS = list(pd.read_csv('/project/LSH/total_data_7727.csv')['ITEMID'].sort_values().unique())

# 1. Indexes of deceased / surviving patients
d_index = np.where(y==1)[0]
s_index = np.where(y==0)[0]

# 2. Separating deceased / surviving patients
d_X = X[d_index]
s_X = X[s_index]

result = []
result_s = []

for d in range(10):
    for f in range(d_X.shape[-1]):
        d_sum = d_X[:,d,f].sum()/d_X.shape[0]
        s_sum = s_X[:,d,f].sum()/s_X.shape[0]
        result.append({'cols':COLS[f], 'day':10-d,'per':d_sum})
        result_s.append({'cols':COLS[f], 'day':10-d,'per':s_sum})
        
d_df = pd.DataFrame(result).sort_values(['cols','day']).reset_index(drop=True)
s_df = pd.DataFrame(result_s).sort_values(['cols','day']).reset_index(drop=True)
d_df.shape, s_df.shape

def absum(top10_list):
    plt.figure(figsize = (12,10), dpi=150)
    i = 0
    for f in top10_list:
        if f == 0:
            continue
        plt.subplot(4,4,1+i)
        plt.title(feature_name[f])
        ax = sns.lineplot(data = d_df[d_df['cols']==int(f)], x = 'day', y='per', label='사망')
        ax = sns.lineplot(data = s_df[s_df['cols']==int(f)], x = 'day', y='per', label='생존', linestyle=':', marker='o')
        ax.invert_xaxis()
        ax.legend(loc='upper left')
        i += 1
    plt.tight_layout()
```

## x_(7727, 10, 3595)

```python
# 1) DATA 
x = np.load('x_(7727,10,3595).npy')
y = np.load('/project/LSH/y_(7727,1).npy')

total_data = pd.read_csv('total_data_7727_10_3595.csv')
features = total_data['ITEMID'].sort_values().unique()

# 2) MODEL
from keras.models import load_model
lstm2 = load_model('./model/f3595-06-val_loss:0.5522.hdf5') # Model for 3595 features

results = []
with tf.device('/device:GPU:1'):
    
    for i in tqdm(range(len(features))):

        save_col = x[:,:,i].copy()

        # 3) Distant Time : E(0to1) - E(1to0)
        x[:,:5,i] = 1
        원_mean_0to1 = np.mean(lstm2.predict(x, batch_size=10000, workers=-1, use_multiprocessing=True))
        x[:,:5,i] = 0
        원_mean_1to0 = np.mean(lstm2.predict(x, batch_size=10000, workers=-1, use_multiprocessing=True))
        
        x[:,:,i] = save_col
        
        # 4) Recent Time : E(0to1) - E(1to0)
        x[:,5:,i] = 1
        근_mean_0to1 = np.mean(lstm2.predict(x, batch_size=10000, workers=-1, use_multiprocessing=True))
        x[:,5:,i] = 0
        근_mean_1to0 = np.mean(lstm2.predict(x, batch_size=10000, workers=-1, use_multiprocessing=True))
        
        x[:,:,i] = save_col
        
        results.append({'feature':features[i], 'distant_mean_diff':원_mean_0to1-원_mean_1to0, '근_mean_diff':근_mean_0to1-근_mean_1to0})
```

```python
df = pd.DataFrame(results)
df.to_csv('m2_df_3595.csv', index = False)
```

```python
# 1. Calculating Entropy -> 'entropy_3595.csv'

from scipy.stats import entropy

x = np.load('x_(7727,10,3595).npy')
y = np.load('/project/LSH/y_(7727,1).npy')

a = pd.read_csv('total_data_7727_10_3595.csv')
features = list(a['ITEMID'].sort_values().unique()) # 3595 ITEMIDs 

results = []
for k in tqdm(range(len(features))):
    rate_1 = x[:,:,k].sum()/(7727*10)
    rate_0 = 1-rate_1
    k_entropy = entropy(np.array([rate_1, rate_0]), base = 2)
    
    results.append({'feature':features[k],'entropy':k_entropy})
    
entropy_df = pd.DataFrame(results)
entropy_df.to_csv('entropy_3595.csv', index = False)
```

# M2_Special Patterns

```python
# DataFrame to be used
# 'distant_mean_diff': difference between mean when changing from 0 to 1 and mean when changing from 1 to 0 for distant time
# 'recent_mean_diff': difference between mean when changing from 0 to 1 and mean when changing from 1 to 0 for recent time

df1 = pd.read_csv('m2_df_3595.csv')
en_df = pd.read_csv('entropy_3595.csv')

m2_3595_score = pd.merge(df1, en_df, on='feature')
m2_3595_score['FI_distance'] = m2_3595_score['distant_mean_diff'] * m2_3595_score['entropy']
m2_3595_score['FI_proximity'] = m2_3595_score['recent_mean_diff'] * m2_3595_score['entropy']
m2_3595_score['Sign'] = m2_3595_score['FI_proximity'].apply(lambda x: 1 if x >= 0 else -1)
m2_3595_score['FI_inversion'] = m2_3595_score['FI_distance'] * m2_3595_score['FI_proximity']
m2_3595_score
```

## Distant Time - No Marker Function / Recent Time - Death Marker

```python
# 1) Recent Time - Death Marker
df1 = m2_3595_score.copy()
df1 = df1[['feature', 'distant_mean_diff', 'recent_mean_diff', 'entropy']]
df1['term1'] = np.exp(-abs(df1['distant_mean_diff'])) ** 10000
df1['term2'] = df1['recent_mean_diff']
df1['term3'] = df1['entropy']
df1['Recent_Time_Death_Marker'] = df1['term1'] * df1['term2'] * df1['term3']

df1 = df1.sort_values(by = 'Recent_Time_Death_Marker', ascending = False)
df1[:10]
absum(df1[:10]['feature'])
## D5W, NS, sodium
```

## Distant Time - No Marker Function / Recent Time - Survival Marker

```python
# 1) Recent Time - Death Marker
df1 = m2_3595_score.copy()
df1 = df1[['feature', 'distant_mean_diff', 'recent_mean_diff', 'entropy']]
df1['term1'] = np.exp(-abs(df1['distant_mean_diff'])) ** 10000
df1['term2'] = df1['recent_mean_diff']
df1['term3'] = df1['entropy']
df1['Recent_Time_Death_Marker'] = df1['term1'] * df1['term2'] * df1['term3']

df1['Recent_Time_Survival_Marker'] = -(df1['Recent_Time_Death_Marker'])
df1 = df1.sort_values(by = 'Recent_Time_Survival_Marker', ascending = False)

df1[:10]
absum(df1[:10]['feature'])

## multivitamins, citalopram, sertraline
```

## Distant Time - Survival Marker / Recent Time - Death Marker

```python
# Definition of the ReLU Function
def relu(x):
    return np.maximum(0, x)

feature_df = pd.read_csv('feature_df.csv')
feature_name = dict(zip(feature_df['feature'], feature_df['feature_name']))

# DataFrame to be used
# 'distant_mean_diff': Difference between the mean when changing from 0 to 1 and the mean when changing from 1 to 0 for distant time
# 'recent_mean_diff': Difference between the mean when changing from 0 to 1 and the mean when changing from 1 to 0 for recent time

df1 = pd.read_csv('m2_df_3595.csv')
en_df = pd.read_csv('entropy_3595.csv')

m2_3595_score = pd.merge(df1, en_df, on='feature')
m2_3595_score['FI_distance'] = m2_3595_score['distant_mean_diff'] * m2_3595_score['entropy']
m2_3595_score['FI_proximity'] = m2_3595_score['recent_mean_diff'] * m2_3595_score['entropy']
m2_3595_score['Sign'] = m2_3595_score['FI_proximity'].apply(lambda x: 1 if x >= 0 else -1)
m2_3595_score['FI_inversion'] = m2_3595_score['FI_distance'] * m2_3595_score['FI_proximity']
m2_3595_score['ITEM_NAME'] = m2_3595_score['feature'].map(feature_name)
m2_3595_score = m2_3595_score[['feature', 'ITEM_NAME', 'distant_mean_diff', 'recent_mean_diff', 'entropy']]
```

```python
df1 = m2_3595_score.copy()
df1['mfar0-mfar1'] = -df1['distant_mean_diff']
df1['DistantSurvival_RecentDeath'] = relu(df1['mfar0-mfar1']) * relu(df1['recent_mean_diff']) * df1['entropy']
df1 = df1.sort_values(by = 'DistantSurvival_RecentDeath', ascending = False)
df1[:10]
absum(df1[:10]['feature'])

# calcium, glucose, arterial line
```

## Distant Time - Death Marker / Recent Time - Survival Marker

```python
df1 = m2_3595_score.copy()
df1['mnear0-mnear1'] = -df1['recent_mean_diff']
df1['DistantDeath_RecentSurvival'] = relu(df1['distant_mean_diff']) * relu(df1['mnear0-mnear1']) * df1['entropy']
df1['DistantDeath_RecentSurvival'].value_counts()
df1 = df1.sort_values(by = 'DistantDeath_RecentSurvival', ascending = False)
df1[:10]
absum(df1[:10]['feature'])

## calcium carbonate, Famotidine, Phenytoin
```
